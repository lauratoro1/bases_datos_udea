## 1. Selección de variables

```python
from pymongo import MongoClient
import pandas as pd
```

```python
# Conexión MongoDB
cliente = MongoClient("mongodb://localhost:27017/")

# Base de datos
db = cliente["taller4_db"]

# Colección
coleccion = db["raw_data"]
```

```python
# Leer documentos MongoDB
datos = list(coleccion.find())

# Ver cantidad de documentos
print(f"Total documentos: {len(datos)}")
```

```python
# Crear DataFrame seleccionando variables relevantes
df = pd.DataFrame(datos)[[
    "id",
    "firstName",
    "lastName",
    "fullName",
    "title",
    "family",
    "imageUrl"
]]

# Mostrar primeras filas
df.head()
```

## 2. Análisis exploratorio

# Inspección 

```python
# Primeras filas del DataFrame
df.head()
```

```python
# Información general del DataFrame
df.info()
```

```python
# Estadísticas descriptivas
df.describe(include="all")
```

# Insights numericos

```python
# Total de personajes
print("Total personajes:", df.shape[0])
```

```python
# Familias únicas
print("Cantidad de familias únicas:", df["family"].nunique())
```

```python
# Familia con más personajes
print(df["family"].value_counts().head(1))
```

```python
# Longitud del título
df["longitud_title"] = df["title"].astype(str).apply(len)

personaje_titulo_largo = df.loc[df["longitud_title"].idxmax()]

print("Personaje con el título más largo:")
print(personaje_titulo_largo["fullName"])
print(personaje_titulo_largo["title"])
```

```python
# Promedio longitud nombres completos
df["longitud_nombre"] = df["fullName"].apply(len)

print(
    "Promedio longitud nombres:",
    round(df["longitud_nombre"].mean(), 2)
)
```

# visualización

```python
import matplotlib.pyplot as plt
```

```python
# Top 5 familias
familias = df["family"].value_counts().head(5)

# Pie chart
plt.figure(figsize=(8,8))

plt.pie(
    familias,
    labels=familias.index,
    autopct='%1.1f%%'
)

plt.title("Distribucion de personajes por familia")

plt.show()
```

```python
# Bar chart
familias.plot(
    kind="bar"
)

plt.title("Top 5 familias con mas personajes")

plt.xlabel("Familia")

plt.ylabel("Cantidad de personajes")

plt.xticks(rotation=45)

plt.show()
```

```python
# Crear variable longitud nombre
df["longitud_nombre"] = df["fullName"].apply(len)

# Histograma
plt.hist(df["longitud_nombre"])

plt.title("Distribucion de longitud de nombres")

plt.xlabel("Cantidad de caracteres")

plt.ylabel("Frecuencia")

plt.show()
```
